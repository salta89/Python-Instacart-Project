### Grouping Data 

```python
# Splitting data into groups based on product name
df.groupby('product_name')
```

### Aggregating Data with agg() 

#### Performing single aggregation 

```python
# Splitting data into groups based on department_id
df.groupby('department_id').agg ({'order_number':['mean']})
```

```python
df.groupby('department_id') ['order_number'].mean()
```

#### Performing multiple aggregations

```python
df.groupby('department_id').agg ({'order_number':['mean', 'min', 'max']})
```

### Aggregating Data with transform()

```python
#Creating max_order column,grouping by user_id, applying transform function on order_number
df_ords_prods_merge['max_order']=df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

### Step 6 Create a spending flag for each user based on the average price across all their orders 

```python
# Creating mean_spend column 
df_ords_prods_merge['mean_spend'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.head()
```

```python
# Creating column with spender flag based on criteria
df_ords_prods_merge.loc[df_ords_prods_merge['mean_spend'] > 10, 'spender_flag']='Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_spend'] <= 10, 'spender_flag']='High spender'
```

```python
df_ords_prods_merge['spender_flag'].value_counts(dropna=False)
```

```python
df_ords_prods_merge[['user_id', 'spender_flag', 'loyalty_flag']].head()
```

### Step 7 Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column. 

```python
# Creating median_order_frequency column
df_ords_prods_merge['median_order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_order_frequency'] > 20, 'order_frequency_flag']='Non-frequent custome'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_order_frequency'] > 10) & (df_ords_prods_merge['median_order_frequency'] <= 20), 'order_frequency_flag']='Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_order_frequency'] <= 10, 'order_frequency_flag']='Frequent custome'
```

```python
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```
