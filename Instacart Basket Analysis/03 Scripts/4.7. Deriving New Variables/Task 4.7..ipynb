```python
# Define a function 
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#Apply the function 
df['price_range'] = df.apply(price_label, axis=1)
```

### Step 4. Create a new column containing “Most orders,” “Average orders,” and “Fewest orders” labels called “busiest_period_of_day.” 

```python
#Finding hours of day ordered frequency 
df_ords_prods_merge['hour_of_day_ordered'].value_counts(dropna=False)
```

I decided to split value counts into equal thirds for the following labels: "Most orders" = 10, 11, 14, 15, 13, 12, 16, 9; "Fewest orders" = 23, 6, 0, 1, 5, 2, 4, 3; "Average orders" = remaining values. 

```python
#Using for-loop to determine busiest peroid of the day 
result_2=[]

for value in df_ords_prods_merge ['hour_of_day_ordered']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_2.append('Most orders')
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result_2.append('Fewest orders')
    else:
        result_2.append('Average orders')
```

```python
result_2
```

```python
#Creating new column busiest_peroid_of_the_day
df_ords_prods_merge['busiest_peroid_of_the_day']=result_2
```
